```python
import pandas as pd

df = pd.DataFrame()
```

```python
for pull in business_pulls:
  df = df.append(pd.DataFrame(pull), ignore_index=True)

print("shape of df is", df.shape)
```

```python
df.insert(1, "insertedAtTimestamp", 0)
```

```python
df.columns
```

```python
df_mod = pd.DataFrame(columns=df.columns)
```

```python
df_mod.shape
```

```python
import datetime
from time import sleep
# this cell removes duplicate rows and inserts a timestamp while filling out a new df_mod
seen = {}
num_skipped = 0
for i in range(len(df)):
  if df.loc[i, 'alias'] in seen:
    num_skipped += 1
    continue
  df_mod = df_mod.append(df.iloc[i], ignore_index=True)
  df_mod.loc[len(seen), "insertedAtTimestamp"] = datetime.datetime.now()
  seen[df.loc[i, 'alias']] = 1
  sleep(.001)
print(num_skipped)
```

```python
df_mod.shape
```

```python
df_mod = df_mod.drop(columns=['is_closed', 'phone', 'display_phone', 'distance', 'price', 'url', 'image_url', 'transactions'])
```
